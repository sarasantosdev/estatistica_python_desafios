# Atividade - Distribuição de Frequência

```python
import pandas as pd
import numpy as np

np.random.seed(123)

estilos_musicais = ["Rock", "Pop", "Eletrônica", "Funk", "Sertanejo"]
regioes = ["Centro-Oeste", "Nordeste", "Norte", "Sudeste", "Sul"]

dados = pd.DataFrame({
    'Estilo_Musical': np.random.choice(estilos_musicais, 500, replace=True),
    'Regiao': np.random.choice(regioes, 500, replace=True)
})

dados.head()
```

**Gerando a tabela cruzada com as frequências absolutas**

```python
tab_freq_absoluta = pd.crosstab(dados['Estilo_Musical'], dados['Regiao'])

tab_freq_absoluta
```

**Gerando a tabela de frequências relativas por coluna**

```python
tab_estilo_musical_relativa = pd.crosstab(
    dados['Estilo_Musical'], dados['Regiao'], normalize='columns'
)*100

tab_estilo_musical_relativa = round(tab_estilo_musical_relativa,1)

tab_estilo_musical_relativa
```

**Gerando a tabela de frequências relativas por linha**

```python
tab_estilo_musical_relativa_linhas = pd.crosstab(
    dados['Estilo_Musical'], dados['Regiao'], normalize='index'
)*100

resultado = round(tab_estilo_musical_relativa_linhas, 1)

resultado
```

**Gerando a tabela de frequências relativas por total geral**

```python
tab_estilo_musical_relativa_total = pd.crosstab(
    dados['Estilo_Musical'], dados['Regiao'], normalize='all'
)*100

resultado = round(tab_estilo_musical_relativa_total, 1)

resultado
```
