**Transforme as variáveis Sexo, Cor e Anos.de.Estudo em categorical e observe o resultado: crie colunas que tratam as variáveis categóricas nominais e ordinais de dados numéricos. Siga as instruções e dados trazidos no documento. Leia no final a nova tabela com os dados transformados.**

```python
df['Genero'] = (pd.Categorical(df['Sexo'], categories=[0,1], ordered=True).rename_categories(['Masculino', 'Feminino']))

df[['Sexo', 'Genero']].drop_duplicates()
```

```python
df['Raca'] = (pd.Categorical(df['Cor'], categories=[0, 2, 4, 6, 8], ordered=True).rename_categories(['Indígena', 'Branca', 'Preta', 'Amarela', 'Parda']))

df[['Cor', 'Raca']].drop_duplicates()
```

```python
anos_de_estudo = {1:'Sem instrução e menos de 1 ano', 2:'1 ano', 3:'2 anos', 4:'3 anos', 5:'4 anos', 6:'5 anos',
    7:'6 anos', 8:'7 anos', 9:'8 anos', 10:'9 anos', 11:'10 anos', 12:'11 anos', 13:'12 anos',14:'13 anos',
    15:'14 anos', 16:'15 anos ou mais', 17:'Não determinados'
}

anos = sorted(df['Anos.de.Estudo'].unique())

df['Escolaridade'] = pd.Categorical(df['Anos.de.Estudo'], anos, ordered=True )

df['Escolaridade'] = df['Escolaridade'].map(anos_de_estudo)

df[['Anos.de.Estudo', 'Escolaridade']].drop_duplicates()
```

```python
df.head()
```

**Apresente em texto a menor e maior Renda da base de dados: Utilize a função print juntamente a formatação de dados f-string como explicada no documento para exibir estes dados.**

```python
print(f'A maior renda é {df['Renda'].max():,.2f} e  a menor renda é {df['Renda'].min():,.2f}')
```

## Case 2: identificar o perfil das pessoas responsáveis pelo domicílio

**Construa uma tabela de frequências das pessoas responsáveis pelo domicílio por Genero. Adicione também uma coluna com esse valor em porcentagem**

```python
freq_pessoas = (
    df.groupby('Genero', observed=False)
    .size()
    .reset_index(name='freq_absoluta')
    .sort_values(by='Genero', ascending=False)
)
```

```python
freq_pessoas
```

```python
freq_pessoas['freq_relativa'] = round(
    (freq_pessoas['freq_absoluta']/freq_pessoas['freq_absoluta'].sum())*100,1
)
```

```python
freq_pessoas
```

```python
freq_pessoas.columns = ['Sexo Biológico', 'Quantidade', 'Porcentagem (%)']

freq_pessoas
```

**Construa uma tabela de frequências absolutas e outra de relativas cruzando as variáveis Genero e Raca das pessoas responsáveis pelo domicílio.**

```python
tab_pessoas_raca_abs = pd.crosstab(df['Genero'], df['Raca'])

tab_pessoas_raca_abs
```

```python
tab_pessoas_raca_rel = pd.crosstab(df['Genero'], df['Raca'],normalize='all') * 100

resultado = round(tab_pessoas_raca_rel, 2)

resultado
```

**Construa uma tabela cruzada para calcular a Renda média das pessoas responsáveis pelo domicílio em relação ao Genero e Raca**

```python
renda_media = pd.crosstab(df['Genero'], df['Raca'],values=df['Renda'], aggfunc='mean')

resultado = round(renda_media, 2)

resultado
```
